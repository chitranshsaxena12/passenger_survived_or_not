# Data Cleaning

We want to fill in missing age data instead of just dropping the missing age data row. One way to do
this is by filling in the mean age of all the passengers. However we can be smarter about this and 
check the average age by passenger class. for example:

```python
plt.figure(figsize=(12,7))
sns.boxplot(x="Pclass",y="Age",data=main_data,palette="winter")
```

```python
def impute_age(cols):
    Age = cols[0]
    Pclass = cols[1]
    
    if pd.isnull(Age):
            
            if Pclass == 1:
                return 37
            elif Pclass == 2:
                return 29
            else:
                return 24
    else:
        return Age
    
```

Now apply that function

```python
main_data["Age"] = main_data[["Age","Pclass"]].apply(impute_age,axis=1)
```

Now let's check that heat map again !

```python
sns.heatmap(main_data.isnull(),yticklabels=False,cbar=False,cmap="viridis")
```

```python
main_data
```

In sex column change all the male and female with 0 and 1

```python
main_data.replace(to_replace ="male",value = 0,inplace=True)
main_data.replace(to_replace ="female",value = 1,inplace=True)
```

Divide data into x (independent variable) and y (dependent variable) 

```python
x = main_data.drop("Survived",axis=1)
y = main_data["Survived"]
```

Now we are going to split our data into train and test data

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=123)
```

# Fit our model 

```python
decision_model = DecisionTreeClassifier()
decision_model.fit(X_train,y_train)
```

```python
plt.figure(figsize=(20,25))
tree.plot_tree(decision_model,filled=True)
```

```python
y_pred = decision_model.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```

for increasing accuracy use grid search CV

```python
parameter = {'criterion':['gini', 'entropy'],
             "splitter":['best', 'random'],
             "max_depth":[2,3,4,5,6]
    
}
```

```python
clf = GridSearchCV(decision_model, parameter,cv=5)
```

```python
clf.fit(X_train,y_train)
```

```python
clf.best_params_
```

```python
decision_model = DecisionTreeClassifier(criterion= 'gini', max_depth= 6, splitter= 'best')
decision_model.fit(X_train,y_train)
```

```python
y_pred = decision_model.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```

# Evaluation

```python
print(confusion_matrix(y_test, y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
